```python
for node in nodes:
    name = node['name']
    node['race'] = 'Human'
    node['citizenship'] = 'U.S.A.'
    node['gender'] = 'Male'
    node['orientation'] = 'Straight'
    for femaleName in females:
        # If full match OR partly match
        if ((name == femaleName) or ((femaleName in name) and (len(femaleName) > 4))):
            node['gender'] = 'Female'
            break
        # Stacy, Gwen <=> Gwen Stacy
        if ', ' in femaleName: 
            splitted = femaleName.split(', ')
            newName = splitted[1] + ' ' + splitted[0]
            if (newName == name):
                node['gender'] = 'Female'
                break

print('Done')
```
